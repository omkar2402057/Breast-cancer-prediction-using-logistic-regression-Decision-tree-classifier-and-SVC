#Predicting Breast Cancer using LogisticRegression,DecisionTree and Support vector machine


##Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
```

##reading the dataset

```python
dataset=pd.read_csv('breast_cancer.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset.head(3)
```

```python
dataset.info()
```

##Number of benign and malignant observations

```python
benign,malignant = dataset['Class'].value_counts()
print("Number of Benign patients:",benign)
print("Number of Malignant patients:",malignant)

# Create a count plot
plt.figure(figsize=(5, 5))
sns.countplot(x='Class', data=dataset, palette=['blue', 'red'])
plt.title('Number of Benign and Malignant Patients')
plt.xlabel('Class (2 = Benign, 4 = Malignant)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show()
```

##Spliting data into train and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

##Data Normalization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier1=LogisticRegression(random_state=0)
classifier1.fit(X_train,y_train)
```

```python
y_pred=classifier1.predict(sc.transform(X_test))
```

##predicted vs accurate

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

##Visualization(Training and test set)

```python
y = np.where(y == 2, 0, 1)
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Training the logistic regression model on PCA-reduced data
classifier_pca = LogisticRegression(random_state=0)
classifier_pca.fit(X_train_pca, y_train)

# Plotting decision boundary
def plot_decision_boundary(X, y, model, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(('red', 'blue')))
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', cmap=ListedColormap(('red', 'blue')))
    plt.title(title)
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    plt.show()

# Plot decision boundary for training set
plot_decision_boundary(X_train_pca, y_train, classifier_pca, 'Logistic Regression (Training set)')

# Plot decision boundary for test set
plot_decision_boundary(X_test_pca, y_test, classifier_pca, 'Logistic Regression (Test set)')
```

## confusion matrix and classification report

```python
from sklearn.metrics import classification_report,confusion_matrix
lr_cm = confusion_matrix(y_test,y_pred)
lr_cm = pd.DataFrame(lr_cm, columns=['Benign', 'Malignant'], index=['Benign','Malignant'])
print(lr_cm)
print(classification_report(y_test,y_pred))
```

##Accuracy and standard deviation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier1, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
